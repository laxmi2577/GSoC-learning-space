```python
import os
import mesa
import google.generativeai as genai
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.environ.get("GOOGLE_API_KEY") 
genai.configure(api_key=api_key)
generation_model = genai.GenerativeModel('gemini-2.5-flash')

class LLMAgent(mesa.Agent):
    def __init__(self, model):
        super().__init__(model)
        self.wealth = 0

    def step(self):
        prompt = f"You are an agent in an economic simulation. You currently have {self.wealth} units of wealth. Respond with a single, highly dramatic sentence in the style of a 90s Bollywood villain about your financial situation."
        
        response = generation_model.generate_content(prompt)
        print(f"Agent {self.unique_id}: {response.text.strip()}")

class LLMModel(mesa.Model):
    def __init__(self, n):
        super().__init__()
        self.num_agents = n
        LLMAgent.create_agents(model=self, n=n)

    def step(self):
        self.agents.shuffle_do("step")
```

```python
test_model = LLMModel(3)
test_model.step()
```
